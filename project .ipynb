```python
import pandas as pd 
data = pd.read_csv('Match.csv')
print(data)
```

```python
data['Season_Year'].groupby(data.Season_Year).agg('count')
```

```python
t = data[data['Season_Year']==2011]['Season_Year'].groupby(data['match_winner']).count()
t
```

```python
t.index.tolist()
t.values.tolist()
```

```python
import numpy as np
import matplotlib.pyplot as plt
marks = t.values.tolist()
divisions =t.index.tolist()
index = np.arange(len(divisions))

plt.bar(divisions,marks,color="#1319cd",width =0.70)
plt.show()
```
